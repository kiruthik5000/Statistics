# Sample Chi Square Test

```python
from scipy.stats import chisquare
```

```python
# number of students select i type of ice-cream
observed = [25, 30, 20, 25] # [20, 50] -> obse

expected = [25, 25, 25, 25] # [35, 35] -> equal


chi2, p = chisquare(observed, expected)

print(f"Chi-square Value : {chi2}")
print(f"P-value : {p}")


if p < 0.05:
    print("We reject the idea that students like all flavours equally.")
else:
    print("We do not reject the idea students like all flavours")
```

# Gender - Mode of Study

```python
from scipy.stats import chi2_contingency

table = [[20, 30], # male 20 prefer Online, 30 prefer Offline
         [25, 25]] # female 25 Online, 25 prefer offline

ch2, p_value, dof, e = chi2_contingency(table)

print(f"Chi-square{ch2}")
print(f"P-value {p_value}")
print(f"Degrees of Freedom {dof}")
print(f"Expected outcome {e}")
```
