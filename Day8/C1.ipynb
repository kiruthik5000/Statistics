```python
import pandas as pd
import matplotlib.pyplot as plt
import os, sys


df = pd.read_csv('C1.csv')


df['Retention'] = df['left'].map({1: 'Retained', 0: 'Left'})


plt.figure(figsize=(7,5))
salary_retention = df.groupby(['salary', 'Retention']).size().unstack()
salary_retention.plot(kind='bar', figsize=(7,5))

plt.title("Impact of Salary on Employee Retention")
plt.xlabel("Salary Level")
plt.ylabel("Number of Employees")
plt.xticks(rotation=0)
plt.legend(title="Status")
plt.tight_layout()
plt.show()


dept_rate = df.groupby('Department')['left'].mean() * 100
retention_rate = 100 - dept_rate

plt.figure(figsize=(8,5))
retention_rate.plot(kind='bar', color='skyblue')

plt.title("Department-wise Employee Retention Rate (%)")
plt.xlabel("Department")
plt.ylabel("Retention Rate (%)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
