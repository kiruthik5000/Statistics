```python
import numpy as np
from scipy import stats
```

```python
np.random.seed(42)
std_heights = np.random.normal(152, 5, 40) # mean = 152, std = 5, n = 40
std_heights
```

```python
population_mean = 150
population_std = 5
```

```python
sample_size = len(std_heights)
sample_mean = np.mean(std_heights)

z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

print("="*20)
print("ONE-SAMPLE Z - TEST")

print(f"population mean {population_mean}")
print(f"population std {population_std}")
print(f"sample mean {sample_mean}")
print(f"sample size {sample_size}")
print(f"z - score {z_score:.2f}")
print(f"p - value {p_value:.2f}")
print("=" * 20)

if p_value < 0.05:
    print("REJECT Ho: Students are siginificantly taller (We need to change the sampling strategy)")
else:
    print("Accept students are No siginificant differenc (No need to change the sampling strategy)")
```
