```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('C2.xlsx')
df.head()
# df.shape
```

```python
sample_df = df.sample(n=200, random_state=42)
sample_df.shape
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
sns.histplot(df['Height_cm'], kde=True, ax=axes[0])
sns.histplot(sample_df['Height_cm'], kde=True, binrange=(150, 190),binwidth=5, color='orange', ax=axes[1])
plt.show()
```

```python
samples = []
for i in range(5, 40, 10):
    cur_sample = df.sample(i, random_state=42)
    samples.append(cur_sample)
```

```python
fig, axes = plt.subplots(1, 4, figsize=(25, 5))

for i in range(4):
    sns.histplot(samples[i]['Height_cm'], kde=True, binrange=(140, 190), binwidth=5, ax=axes[i])
```

```python
df = pd.read_excel('C2.xlsx')
df.head(), df.shape
```

```python
import numpy as np
```

```python
result = {'height':[]}

for i in range(500):
    mean = df['Height_cm'].sample(n=50,replace=True).mean()
    result['height'].append(mean)
result = pd.DataFrame(result)
result
```

```python
plt.figure(figsize=(7, 5))
sns.histplot(result['height'], kde=True, color='orange')
plt.title("Sample Distribution of Mean (CLT in Action)")
plt.xlabel("Sample mean Height (cm) (n = 50)")
plt.show()
```
