```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('P2.xlsx')
df.head()
df.shape
```

```python
sample = df.sample(n=50, random_state=42)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

sns.histplot(df['Delivery_Time_Hours'], stat='density', kde=True, color='blue', bins=20, ax=axes[0])
sns.histplot(sample['Delivery_Time_Hours'], stat='density',  kde=True, color='orange', bins=20, ax=axes[1])

plt.show()
```

# Practice 3

```python
result = {'Delivery_Time_Hours':[]}

for i in range(200):
    mean = df['Delivery_Time_Hours'].sample(n=50).mean()
    result['Delivery_Time_Hours'].append(mean)
result = pd.DataFrame(result)
result
```

```python
plt.figure(figsize=(7, 5))
sns.histplot(result['Delivery_Time_Hours'], kde=True, color='green')
plt.title("Sample Distribution of Mean (CLT in Action)")
plt.xlabel("Sample mean Delivery Time")
plt.show()
```
