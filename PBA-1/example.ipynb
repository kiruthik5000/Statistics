```python
import pandas as pd
import numpy as np
np.random.seed(42)
```

```python
df = pd.read_excel('e1.xlsx')
df.head()
# df.shape
```

```python
samples = []

for _ in range(500):
    sample = df['Height_cm'].sample(n=50).mean()
    samples.append(sample)
len(samples)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(samples, binwidth=0.5, color='brown', kde=True)
plt.title("Sampling Distribution of Mean (CLT in Action)")
plt.xlabel("Sample Mean Height (Cm) (n=50)")
plt.show()
```
